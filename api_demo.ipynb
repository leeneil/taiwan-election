## API 示範

### summary

第一個要介紹的 API 是 summary 方法，可以以直轄市，區，或者里三個層級來看查詢的範圍的票數總計。當然如果是非六都，也可以用縣、鄉鎮市區、村里三種等級來查詢。

例如如果查詢「台北市」或「臺北市」，則會傳回整個台北市整體的總體票數：

```python
TWE.summary("臺北市")
```

但如果要限縮到某個行政區呢？則可以加上第二個參數，例如：

```python
TWE.summary("台北市", "南港區")
```

最精細可以查詢到「里」，等級：

```python
TWE.summary("台北市", "南港區", "中研里")
```

```python
TWE.stats_by_li("台北市","南港區")
```

這裡要特別補充的是，中選會提供的資料都是精細到「投開票所」等級，也就是說在大多數的都會地區，一個里就會有複數個（通常三或四個）投開票所。本 API 的設計理念是希望讓有興趣的人可以得到「簡化」的高層次資料。如果對最精細的資料有興趣的朋友，可以直接取用 `TWE.vote_data` 即可得到包含「投開票所」等級精細的原始資料。

### stats_by_li

接著要介紹的是 `stats_list_li()`，這是一個可以列出查詢的範圍下，以里為單位的投票資料。和剛才介紹的 `summary()` 一樣，可以以市、行政區、里三個層級查詢，但是不同的是 `summary()` 列出的是查詢範圍的加總資料，而 `list_li` 則是永遠會列出以村里為單位的所有資料。例如如果我們查詢「台北市」，可以看到新北市下面全部的里資料：

```python
TWE.stats_by_li("台北市")
```

```python
TWE.stats_by_li("台北市", "南港區")
```

## 其他輔助函式

目前完成的其實就只有 `summary()` 跟 `stats_by_li()` 兩個比較高階的函式而已。不過下面一些輔助函式可能也會對各位有點幫助。

### list_li

`list_li()` 可以查詢行政區域下有哪些里，也可以純粹拿來數有多少里，哈哈。

```python
TWE.list_li("台北市")
```

```python
TWE.list_li("台北市", "南港區")
```

### list_candidates

另一個可能幫得上忙的輔助函式是 `list_candidates()`，可以印出所有候選人的資料。

```python
TWE.list_candidates("台北市")
```
